# This is an implementation of the Normal Equation

## Equation: AT * A * proj<column space of Ax>(b) = AT * b or proj<column space of Ax> = (AT * A)^-1 * AT * b

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Define some points

```python
num_points = 100
num_weights = 1 #including bias term

# Generate random data
# A = np.random.rand(num_points, num_weights + 1) # random [X, bias]
# b = np.random.rand(num_points, 1) # random target values

# Generate random data with a linear relationship plus some noise
A = np.random.rand(num_points, num_weights + 1)
A[:, 1] = 1  # set bias term
true_weights = np.array([[2], [3]])  # true weights for generating b
b = A @ true_weights + np.random.randn(num_points, 1) * 0.1  # add some noise

plt.plot(A[:, 0], b, 'o')
plt.title('Random Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

## Calculate Least Squares Solution

```python
least_squares_weights = np.linalg.inv(A.T @ A) @ A.T @ b
print("Least Squares Weights:\n", least_squares_weights)
```

## Plot output of Least Square solutions

```python
plt.plot(A[:, 0], b, 'o', label='Data Points')

#plot least squares fit line with only weights for bias and single feature
plt.plot(A[:, 0], A @ least_squares_weights, "o", label='Least Squares Fit')
plt.title('Least Squares Fit using Normal Equation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

## Plot weights of least square solution

```python
plt.plot(A[:, 0], b, 'o', label='Data Points')

x_vals_line = np.linspace(0, 1, 100)
y_vals_line = least_squares_weights[0] * x_vals_line + least_squares_weights[1]
plt.plot(x_vals_line, y_vals_line, 'r-', label='Least Squares Fit')

#show weights on the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
weights_text = f'Weights:\n w1: {least_squares_weights[0][0]:.2f}\n w0: {least_squares_weights[1][0]:.2f}'
plt.text(0.05, 0.95, weights_text, transform=plt.gca().transAxes, fontsize=10,
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))
plt.title('Least Squares Fit using Normal Equation')
plt.grid(True)
plt.show()


```
